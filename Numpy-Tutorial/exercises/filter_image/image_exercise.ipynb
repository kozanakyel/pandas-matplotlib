```python
import numpy as np
```

```python
def smooth(img):
    avg_img =(      img[1:-1, 1:-1]  # center
                  + img[ :-2, 1:-1]  # top
                  + img[2:  , 1:-1]  # bottom
                  + img[1:-1,  :-2]  # left
                  + img[1:-1, 2:  ]  # right
                ) / 50.0
    return avg_img
```

```python
img = plt.imread('dc_metro.png')
avg_img = smooth(img)

plt.figure()
plt.gray()
plt.subplot(1,3,1)
plt.imshow(img)
plt.title("original")

plt.subplot(1,3,2)
plt.imshow(avg_img)
plt.title("smooted once")

plt.subplot(1,3,3)
plt.imshow(img[1:-1,1:-1] - avg_img)
plt.title('difference')
```

```python
for num in range(1000):
    avg_img = smooth(img)
```

```python
# Plot the original image first
plt.figure()
plt.subplot(1,2,1)
plt.imshow(img)
plt.title('original')

# Now the filtered image.
plt.subplot(1,2,2)
plt.imshow(avg_img)
plt.title('smoothed 50 times')
```
