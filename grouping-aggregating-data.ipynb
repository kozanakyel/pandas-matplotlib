```python
country_respondents = df['Country'].value_counts()
country_respondents
```

```python
country_think_agree = country_grp['SeriousWork'].apply(lambda x: x.str.contains('agree').sum())
country_think_agree
```

```python
agree_df = pd.concat([country_respondents, country_think_agree], axis='columns', sort=False)
agree_df
```

```python
agree_df.rename(columns={'Country': 'NumRespond', 'SeriousWork': 'SayAgree'}, inplace=True)
```

```python
agree_df
```

```python
agree_df['PctAgree'] = (agree_df['SayAgree']/agree_df['NumRespond']) * 100
agree_df
```

```python
agree_df.sort_values(by='PctAgree', ascending=False)
agree_df.head(20)
```
